## Exercise 3
a) Create boxplot of apartment's grade versus it's price. Compare with plot 2b).  
b) Create mosaic plot of any two categorical variables.

```python
#a)
#Your code:
grades = [ [] for _ in range(0, 13)]
for i, grade in enumerate(data['grade']):
    grades[grade-1].append(data['price'][i])
    
from pylab import rcParams
rcParams['figure.figsize'] = 7, 5
plt.boxplot(grades)
plt.xlabel('grade')
plt.ylabel('price')
plt.show()

#Hint: plt.boxplot
```

```python
#b)
from statsmodels.graphics.mosaicplot import mosaic
#Your code:

view = data['view']
grade = data['grade']
d = pd.DataFrame({'view': view, 'grade': grade})
mosaic(data,['view','grade'])
plt.show()
```

## Exercise 4
"I am interested in the house with 3 bedrooms, 2 bathrooms and 2 floors"  
a) Compute the basic statistics of apartments fulfilling above constraints.  
b) Distinct accepted apartments on any of previously created plots.  
c) Find such apartments with best price to footage ratio.  

```python
#a)
#Your code:
d = data.copy()
d = d.loc[d['bedrooms'] == 3]
d = d.loc[d['bathrooms'] == 2]
d = d.loc[d['floors'] == 2]
d.describe()
#Hint: pandas.DataFrame.loc
```

```python
#b)
#Your code:
plt.scatter(x=d['sqft_living'], y=d['price'])
plt.xlabel('sqfrt living')
plt.ylabel('price')
```

```python
#c)
#Your code:
d['ratio'] = d.apply(lambda row: row.price / row.sqft_living, axis=1)
d.sort_values(by='ratio')
d.head()

#Hint: .loc, .sort_values(), .index
```
